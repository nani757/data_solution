```python
import pandas as pd
import polars as pl
```

```python


def read_file_pd(file_path):
    if file_path.endswith('.csv'):
        # File is CSV
        df = pd.read_csv(file_path)
        df.info()
    elif file_path.endswith('.xlsx') or file_path.endswith('.xls'):
        # File is Excel
        df = pd.read_excel(file_path)
        df.info()
    else:
        # File format not supported
        print("Unsupported file format. Please provide a CSV or Excel file.")
        return None
    
    return df

# Example usage
file_path = "Shark_Tank_India_Dataset.csv"
# file_path = input("Enter the file path: ")
data_frame = read_file_pd(file_path)

if data_frame is not None:
    print("Data successfully loaded:")
    print(data_frame.head())
```

```python
import pandas as pd
import os

def read_file_pd(file_path):
    if file_path.endswith('.csv'):
        # File is CSV
        df = pd.read_csv(file_path)
        df.info()
    elif file_path.endswith('.xlsx') or file_path.endswith('.xls'):
        # File is Excel
        df = pd.read_excel(file_path)
        df.info()
    else:
        # File format not supported
        print("Unsupported file format. Please provide a CSV or Excel file.")
        return None
    
    return df

def read_file_Pl(file_path):
    if file_path.endswith('.csv'):
        # File is CSV
        df = pl.read_csv(file_path)
        df.info()
    elif file_path.endswith('.xlsx') or file_path.endswith('.xls'):
        # File is Excel
        df = pl.read_excel(file_path)
        df.info()
    else:
        # File format not supported
        print("Unsupported file format. Please provide a CSV or Excel file.")
        return None
    
    return df

# Example usage
file_path = "Shark_Tank_India_Dataset.csv"
# file_path = input("Enter the file path: ")

# Check dataset size
file_size = os.path.getsize(file_path)
if file_size > 100 * 1024 * 1024:  # 100 megabytes
    data_frame = read_file_Pl(file_path)
else: 
    data_frame = read_file_pd(file_path)

if data_frame is not None:
    print("Data successfully loaded:")
    print(data_frame.head())
```
